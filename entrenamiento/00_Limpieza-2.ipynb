```python
#elimnamos registro con fecha en formato raro 0023-12-25 a 23-12-2025
df_reservas_3 = df_reservas_3[df_reservas_3['FECHA_TOMA'] !='23-12-2025 04:35:26.000']

#Elimino la hora de la fecha, no es relevante
df_reservas_3['FECHA_TOMA'] = pd.to_datetime(df_reservas_3['FECHA_TOMA']).dt.date

df_reservas_3["FECHA_TOMA"].value_counts(normalize=True)

df_reservas_3['ANTELACION_DIAS'] = (
    pd.to_datetime(df_reservas_3['LLEGADA']) -
    pd.to_datetime(df_reservas_3['FECHA_TOMA'])
).dt.days

#
#bins = [-1, 7, 30, 90, 180, 365, 9999]
#labels = [
#    'ULTIMA_SEMANA',
#    '1_MES',
#    '3_MESES',
#    '6_MESES',
#    '1_ANO',
#    'MAS_1_ANO'
#]
#df_reservas_3['ANTELACION_TRAMO'] = pd.cut(
#    df_reservas_3['ANTELACION_DIAS'],
#    bins=bins,
#    labels=labels
#)
```

```python
#creo funcion para agrupar en temporadas las fechas de llegada (revisar que es temporada alta/baja en sudamerica)
def temporada_hotel(fecha):

    mes = fecha.month

    if mes in [12,1,2,3,4,5]:
        return 'ALTA'

    elif mes in [6, 10,11]:
        return 'MEDIA'

    else:
        return 'BAJA'

df_reservas_3['TEMPORADA'] = (pd.to_datetime(df_reservas_3['LLEGADA']).apply(temporada_hotel))
```

```python
#Vamos a quedarnos con los paises con mas volumen y agrupar los países minoritarios en una categoría común llamada "OTROS"
#Ademas vamos a rescatar y mantener separados a los países que tengan un comportamiento de cancelación extremo.

# OJO! Tenemos una opcion que es "SIN PAIS"

# Mantenemos los de mucho volumen + los que descubrimos con cancelaciones extremas
paises_estrategicos = [
    'ESTADOS UNIDOS', 'MEXICO', 'CANADA', 'BRASIL', 'ARGENTINA',
    'REPUBLICA DOMINICANA', 'ESPAÑA', 'REINO UNIDO', 'ALEMANIA',
    'SIN PAIS',             # Volumen alto y 61% de cancelación
    'COREA DEL SUR',        # ¡85% de cancelación!
    'AFGANISTAN',           # 59% de cancelación
    'EMIRATOS ARABES UNIDOS' # 59% de cancelación
]

# Agrupamos el resto en 'OTROS'
df_reservas_5['PAIS_AGRUPADO'] = df_reservas_5['PAIS'].apply(lambda x: x if x in paises_estrategicos else 'OTROS')

# Verificamos cómo queda la distribución ahora
df_reservas_5['PAIS_AGRUPADO'].value_counts(normalize=True)
```
